## Introduction:
#### Governments collect and spend revenues taken from the earnings of the populous of people in their jurisdictions. Revenues are typically collected in the form of taxation, though the structure and setting of how collections occur vary across the different types and levels of governments. Types of taxation include sales/consumption, excise, income, property, capital gains, and more. Most often, governments attempt to collect revenues from more than one source to diversify their revenue in-stream for purposes typically including taxation equity and revenue stability. 

#### When governments collect revenue from multiple sources, it can become difficult to observe how much tax is actually being collected. When revenue is collected in the form of sales-taxes, property taxes, income taxes, and more all at the same time, people lose the ability to see how large their tax burden actually is. This can be complicated even further by taxation from various levels of government (i.e. local, state, federal, and other taxing districts).  

#### Governments use this notion to their advantage. When revenue structures are complex, governments at various levels can extract more in taxes from their populations than they believe they are paying. This is the concept of Fiscal Illusion, and is defined as, “… the notion that the systematic misperception of key fiscal parameters may significantly distort fiscal choices by the electorate,” (Oates, 1985). Put simply: the structure of the institutions (tax structures) whereby citizens pay for government (taxes) can have effects on how people perceive the cost of government services as the transaction costs of learning about and understanding how each tax dollar is spent increases. 

#### One of the first and foremost papers in this realm is Wagner (1976). In this paper, Wagner describes and reiterates how taxpayer’s perceptions of the price of government can be influenced by how it is financed. In this article, Wagner attempts to analyze the impact of the simplicity of the revenue structure on total government expenditures. He utilizes tax collection, expenditure, and congruent data from the 50 of the largest U.S. cities, and develops a variable to measure the simplicity of the revenue structure, using a Herfindahl Index. 

#### This paper is based heavily off of Wagner (1976). I use effectively the same model and approach, but extend his study by utilizing 13 years of data in a panel dataset, and by conducting an analysis at the state level rather than the city level to see if the theory still holds. This extension tests the use of both a panel fixed effects regression, as well as a simple pooled OLS regression and other iterations of the model and  utilizes multiple constructions of the index variable to create additional knowledge surrounding this theory analyzed by Wagner over 40 years ago.


## Literature Review:

#### Public Choice Theory is an area of study and framework for analysis within economics. The primary basis of this framework is the use of the tools and techniques of economists and applying them to analyzing the behavior of actors in the public sector. One of the largest underpinnings of this theory relates to the lack of incentive for voters to effectively monitor government (Shaw, n.d.). The basis of this issue extends into one of the theories of Public Finance: the theory of Fiscal Illusion. 

#### The seminal paper on fiscal illusion is Puviani (1903). He presents the concept that the public fiscal structure is an institution whereby the ruling class can extract funds and wealth from the ruled of lower classes. Puviani notes that, to prevent resistance on the part of the dominated class, the elites will organize the fiscal system so to disguise the degree of wealth that is being extracted. They do this in a multitude of ways, including: 1) obscuring the cost of government, 2) disguising expense during periods of large spending, 3) explicit nominal fees, 4) through shifts in public attitude, 5) scare tactics to induce greater taxation, 6) breaking up taxation into multiple, small levies instead of a few large ones, and 7) disguising who pays for the explicit public service. The modern foundational and influential work comes from Buchanan (1960, 1967), who discovered and built on the works of Puviani and other Italian political economists.
#### Buchanan (1967) comments on and writes about the work of Puviani and applies the works of Puviani to democratic settings, whereby he notes that, distinct from in non-democratic societies, individuals must be allowed to “purchase” public goods and services through political institutions in a manner that is as non-distortive as possible. He presents income-withholding and progressive income taxation, among other methods, as examples of applied fiscal illusion in democratic institutions. Wagner (1976) follows with the paper this study is based upon, titled: Revenue Structure, Fiscal Illusion, and Budgetary Choice. 
#### Fundamental to these previously presented frameworks is this Downsian (Downs, 1957) model of politicians’ utility-maximization. He assumes in this model that politicians are self-interested in maximizing the probability of being elected, so they choose to engage in actions and pursue policies that would allow them to gain the broadest level of support. On the flip side, he assumes that local voters are also utility-maximizers that will vote for the politician that will maximize their net-benefit. Thus, the politician uses taxes and expenditure policy to attract voter support with lower taxes and higher expenditures being assumed to be more favorably viewed and supported by voters and taxpayers than higher taxes and lower expenditures. 
#### Paul Courant, Edward Gramlich and Daniel Rubinfeld (1979) and Wallace Oates (1979), in distinct papers develop the argument that politicians that seek to maximize their budget, or maximize the utility of the median voter, will use lump-sum grants to disguise additional spending from taxpayers. In the former paper, the authors also assert that fiscal illusion comes specifically from the inability of voters to distinguish between the average cost that they pay in taxes for local public goods and the true marginal tax cost observed.
#### West and Winer (1980) reiterate the ideas surrounding fiscal illusion expressed in the previous literature, with particular emphasis on the tax perception concept noted in Wagner and Buchanan (1977). The authors use the context of the previous literature to build a model to describe the optimal investments that governments place into resources for disguising the costs of government services, given that government. This is prefaced by the explanation of a model that explains the utility maximization of government, wherein it is assumed that bureaucrats prefer more government resources to less. In this model, West and Winer assume that the government seeks to maximize the median voter utility. The median voter theorem states that: for governments to create a winning coalition, they must formulate policies and programs that appeal to the, “median position of voter preferences.” They note that the governments will seek to appeal to the median voter to maintain power; to do so, governments attempt to provide more public services than the observed cost by the taxpayers for these goods. This leads to governments attempting to engage in fiscal illusion to maximize the observed cost to observed benefit ratio.
#### Largely, the literature concludes that fiscal illusion exists, and governments tend to exploit this to enlarge how much revenue they take in. Politicians may use fiscal illusion to inflate how extensive the public services being offered are relative to expenditures to win elections. This incentive may lead to increases in revenue structure complexity to hide these increases in revenue collections. This is the basis of the theory I use in framing my analysis.


## Theory, Data and Model:

### Theory:
#### The theory I use to frame this study is functionally identical to the one proposed in Wagner (1976). He writes: 
#### "In the simplest fiscal setting, only one type of FED (Fiscal Extraction Devices) exists, and it operates openly and clearly. In a complex fiscal setting, FEDs are numerous, are often inconspicuously placed, and sometimes operate according to poorly defined rules […] The formation of an accurate perception regarding the price of public output would be vastly more difficult under this more complex revenue structure."
#### When there are more FED’s (or mechanisms of taxation/revenue gathering), the public likely greater difficulty deciphering how much they pay. This can be considered an increase in transaction costs to observing the cost of government services by the taxpayer. This increase in transactions costs of observing how much individuals pay for services is likely to lead to an underestimation of the total tax bill. He thus hypothesizes that entities that resort to a more complex revenue structure will lead to a growth in public-sector expenditure. I hypothesize and theorize the same.
#### His analysis incorporated data from 50 of the largest cities, excluding those that had dependent school districts. I extend this analysis to state government revenues and expenditures to determine if his findings, which validated the hypothesis he proposed, hold at the state level of analysis. 


### Model: 
#### The model utilized by Wagner primarily revolved around the development of a Herfindahl Index to provide cities with a score of their revenue complexity. A Herfindahl Index attempts to measure the level of concentration in a market (Hayes, 2021). In this instance, Wagner applied it to the revenue structure of cities to determine the concentration, or the simplicity of the revenue structure. This score was constructed in the same manner as a Herfindahl Index is built in other settings in economics, like in industrial organization. This can be denoted mathematically as follows: 
#### ${S}= \sum \limits_{n=1} ^{p} R_{i}^{2}$
#### Where S is the simplicity score of given to the city’s revenue structure, and R_i^2 is the share of the total city’s revenue generated by a particular form of taxation or monetary extraction. The rest of the independent variables Wagner uses in the model are as follows: 
#### X1: total personal income, X2: intergovernmental revenue, X3: percentage of population below poverty line, X4: average salary of city employees, X5: local expenditure as a percentage of state and local expenditure within the state, X6: city population as a percentage of SMSA population, X7: population density, and X8: simplicity of the city’s revenue structure (S, as mathematically displayed above).  
#### These variables were included due to their perceived impact on total city expenditure and their understood impact on being determinants of public expenditure, per the public finance literature. The dependent variable used in his model is total current city expenditure. I apply functionally the same model as the model translates well to the state level of analysis. As such, state level data for all 50 state is collected and used. 

#### As state revenue structures are more complex, I developed five Herfindahl Indices: one that has all 8 forms of revenue at the state level as categorized by the data, and one that has only 5, accounting for only the methods of revenue collection that are forms of explicit taxation. This was done as I identified these two separate ways in which the revenues could be categorized; all revenue and only revenue from taxes. The lower the number of factors, the greater the variance is likely to be due to how the Herfindahl index formula calculates the value of the variable. The index score with only tax collections was composed of: 1) sales taxes, 2) property taxes, 3) individual income taxes, 4) corporate taxes, and 5) other taxes. The index with all 8 components included all the component of the other index in addition to: 1) current charges, 2) miscellaneous general revenue, and 3) utility revenue. I constructed three additional variables that combined some of factors. The first of these I named, "ndex Score Grouped, Utilities Included" and it grouped Individual, Corporate, and Other taxes into one factor (i.e. this value was the sum of these components, divided by two and squared), as well as the miscellanous and charges into one and included utility revenue. The second of these, titled, "Index Score Grouped, No Utilities Included," grouped the variables as in the previous, but excluded utilities (where the values were divided by te total revenue in the state minus utilities). The final one of these is the same as the prior but included intergovernmental revenue from the federal government as a separate component in the calculation of the index.

#### Intergovernmental transfers were left out of the construction of the index as they are included as another independent variable in the model. The dependent variable in my model is the total State Government Expenditures per Person. The rest of the control variables are as follows: 
#### X1: Proportion of Expenditures in the State coming from the State Government, X2: Total Number of State Employees, X3: Average Wage of State Employees, X4: State Per Capita Income, X5: Proportion of Revenue from Intergovernmental Transfers (from the Fed. Govt.), X7: Proportion of the Population under the Poverty Line, and X8: The Partisan Makeup of the State Legislature. 
#### These specific variables were included as many are the variables most closely in-line with those used by Wagner in his paper, and the others are factors that are observable at the state level that may also influence the degree of financing of state governments. These include the per capita income variable, as well as the partisan makeup variable, as wealthier states, such as New York and California are assumed to have higher state expenditures per person, and the partisan makeup of these state legislatures likely largely impacts the state revenue collection and expenditures as well.

## Conclusion and Discussion:

#### The results from the empirical analysis tell us a few things about the effect of state government level revenue complexity on state revenue and expenditures. The first conclusion that we can draw from these results is that the method of construction of the index scores has a significant impact on the score given to each state, as well as the significance and the magnitude of the coefficient that the given score has in each model. 

#### The second primary conclusion that can be drawn from these results is that, without controlling for a number of factors of variation in the model, there would be inconclusive evidence that the index scores constructed have a statistically signifcant, consistent, negative impact on per capita state expenditures. 

#### Thirdly, there is evidence to suggest that an increase in the revenue simplicity in a given state leads to declines in state government expenditures on a per capita basis. This is not consistent, though for state government revenues per capita, where results are too sporadic to conclude anything with any real degree of confidence.

#### Thus, though these results are not perfectly explanatory, we can say with at least some degree of confidence that Wagner's findings can be extended to the state level of analysis, so long as adaptations are made to to properly fit to the complexities of state governmenta finances over the stretch of 13 years that may not be present in the finances of local and city governments as observed in one year by Wagner.
